```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the webpage containing Tesla's revenue data
url = 'https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue'

# Send a GET request to the webpage
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table element containing the revenue data
table = soup.find('table', class_='historical_data_table')

# Extract the revenue data from the table
revenue_data = []
rows = table.find_all('tr')
for row in rows[1:]:
    cells = row.find_all('td')
    date = cells[0].text.strip()
    revenue = cells[1].text.strip().replace('$', '').replace(',', '')
    revenue_data.append({'Date': date, 'Revenue': revenue})

# Display the extracted revenue data
for data in revenue_data:
    print(data)
```
